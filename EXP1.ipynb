```python
import math
import matplotlib.pyplot as plt
```

```python
def MC(TA, TC, Cell_Area, N):
    Cell_Cap = TC / N
    Cluster_Area = N * Cell_Area
    m = TA / Cluster_Area
    SC = m * N * Cell_Cap
    return SC
```

```python
SC = MC(4200, 1001, 12, 7)
print(SC)
```

```python
N = [1, 2, 3, 4, 5, 6, 7]
SC = [MC(4200, 1001, 12, n) for n in N]

plt.plot(N, SC)
plt.title('Graph of N vs SC')
plt.xlabel('N')
plt.ylabel('SC')
plt.show()
```
