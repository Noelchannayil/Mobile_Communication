```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Exp2(N):
    Q=math.sqrt(3*N)
    return Q
```

```python
Q=Exp2(4)
print(Q)
```

```python
n1=[]
n2=[]
for i in range(1,5):
  for j in range(0,5):
    N=i**2+j**2+i*j
    n1.append(N)
    Q=Exp2(N)
    n2.append(Q)
print(n1)
print(n2)
```

```python
plt.stem(n1,n2)
plt.title('Effect of cluster size on signal to interference ratio')
plt.grid()
plt.xlabel('N')
plt.ylabel('Q')
plt.show()
```

```python
def Exp2(N,n,io):
  Q=math.sqrt(3*N)
  S_I= 10*math.log10((Q**n)/io)
  return S_I
```

```python
for i in range(1,5):
  for j in range(1,5):
    N=i**2+j**2+i*j
    if N in Nstd:
      e=N
      Q=Exp2(e,4,6)
      S_I=Exp2(e,4,6)
      print('SI is =',S_I)
      print('Q is =',Q)
```

```python
# Store the values of N and S_I in lists as they are calculated.
N_list = []
S_I_list = []
for i in range(1,5):
  for j in range(1,5):
    N=i**2+j**2+i*j
    if N in Nstd:  # Assuming Nstd is defined somewhere
      e=N
      Q=Exp2(e,4,6)
      S_I=Exp2(e,4,6)
      N_list.append(N)
      S_I_list.append(S_I)
      print('SI is =',S_I)
      print('Q is =',Q)

# Plot the collected data
plt.stem(N_list,S_I_list)
plt.title('Effect of cluster size on signal to interference ratio')
plt.grid()
plt.xlabel('N')
plt.ylabel('S_I')
# Annotate the plot using the lists
for i in range(len(N_list)):
    plt.annotate(f"({N_list[i]:.2f}, {S_I_list[i]:.2f})", (float(N_list[i]), float(S_I_list[i])), textcoords="offset points",xytext=(0,10), ha='center')
plt.show()
```
