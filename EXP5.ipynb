```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
Pt=4 #Watts
Pl_d0=-32 #dBm
n=4
d=3*(10**3)
d0=100
X_sigma=10.5 #dBm
```

```python
Pt_dBm=10*np.log10(Pt *( 10**3))
print("Pt (dBm) = ",Pt_dBm)
```

```python
PL1=Pl_d0+n*10*np.log10(d/d0)
print("PL1 = ",PL1)
```

```python
PL2=Pl_d0+n*10*np.log10(d/d0) + X_sigma
print("PL2 = ", PL2)
```

```python
Pr2=Pt_dBm-PL2
print("Pr2 = ",Pr2)
```

```python
R1=[Pt_dBm -(Pl_d0 + 10*n*math.log(d/d0,10)) for n in range(1,11)]
R2=[Pt_dBm -(Pl_d0 + 10*n*math.log(d/d0,10)) + X_sigma for n in range(1,11)]
plt.title("n vs Pr")
plt.plot(list(range(1,11)),R1,color='red')
plt.plot(list(range(1,11)),R2,color='blue')
plt.xlabel("n")
plt.ylabel("Pr")
plt.title="Pr vs n"
plt.grid(True)
plt.show()
```
