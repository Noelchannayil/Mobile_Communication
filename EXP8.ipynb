```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def capacity_cal_mimo(SNR,M):
  SNR_ratio = 10 ** (SNR / 10)
  Cap_mimo=M*math.log2(1+SNR_ratio)
  return Cap_mimo
```

```python
def capacity_cal_simo(SNR,M):
  SNR_ratio = 10 ** (SNR / 10)
  cap_simo=math.log2(1+SNR_ratio)
  return cap_simo
```

```python
def capacity_cal_siso(SNR):
  SNR_ratio = 10 ** (SNR / 10)
  cap_siso=math.log2(1+SNR_ratio)
  return cap_siso
```

```python
SNR=[]
Capa_mimo=[]
capa_simo=[]
capa_siso=[]
for i in range(0,51,5):
  SNR.append(i)
  Capa_mimo.append(capacity_cal_mimo(i,2))
  capa_simo.append(capacity_cal_simo(i,2))
  capa_siso.append(capacity_cal_siso(i))
print(SNR)
print(Capa_mimo)
```

```python
plt.plot(SNR,Capa_mimo,label='MIMO')
plt.plot(SNR,capa_simo,label='SIMO')
plt.plot(SNR,capa_siso,label='SISO')
plt.xlabel('SNR(dB)')
plt.ylabel('Capacity(bps)')
plt.title('Capacity vs SNR')
plt.grid()
plt.legend()
plt.xticks(SNR)
plt.show()
```

```python
SNR = list(range(0,51,5))
M_values = list(range(2,11))

#Create a plot
plt.figure(figsize=(10,6))
```

```python
#calculate and plot capacity for each M
for M in M_values:
  Capa_mimo=[capacity_cal_mimo(snr,M) for snr in SNR]
  plt.plot(SNR,Capa_mimo,label=f'M={M}',marker='o')

#Plot formatting
plt.title('MIMO Capacity vs SNR for Different M values')
plt.xlabel('SNR(dB)')
plt.ylabel('MIMO Capacity(bits/s/Hz)')
plt.legend()
plt.grid()
plt.xticks(SNR)
plt.show()
```
