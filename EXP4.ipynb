```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def r_freq(c,Vm,fc,thetha):
  lamb=c/fc
  if (thetha==0):
    fd=(Vm*np.cos(thetha))/lamb
    fr=fc+fd
  elif (thetha==math.pi):
    fd=(Vm*np.cos(thetha))/lamb
    fr=fc+fd
  elif (thetha==math.pi/2):
    fr=fc
  else:
    fd=(Vm*np.cos(thetha))/lamb
    fr=fc+fd
  return fr/10**6
```

```python
fc=900*10**6
c=3*10**8
Vm=72
thetha=0
fr=r_freq(c,Vm,fc,thetha)
print(fr,"Mhz")
thetha1=math.pi
fr1=r_freq(c,Vm,fc,thetha1)
print(fr1,"Mhz")
thetha2=math.pi/2
fr2=r_freq(c,Vm,fc,thetha2)
print(fr2,"Mhz")
thetha3=math.pi/3
fr3=r_freq(c,Vm,fc,thetha3)
print(fr3,"Mhz")
```

```python
fd_values = []
thetha_values = [0,90,180,360]
fr_values=[]
c = 3*10**8
fc = 900*10**6
Vm = 72
```

```python
for thetha in thetha_values:
  thetha_rad = math.radians(thetha)
  lamb = c / fc
  fd = (Vm * math.cos(thetha_rad)) / lamb
  fd_values.append(fd)
  fr=fc+fd
  fr_values.append(fr/10**6)
```

```python
print(fd_values)
print(fr_values)
```

```python
plt.plot(thetha_values,fd_values)
```

```python
plt.plot(fd_values,fr_values)
```
